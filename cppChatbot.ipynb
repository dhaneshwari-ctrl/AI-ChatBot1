```python
from langchain_community.document_loaders import TextLoader
from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS

# ✅ Force UTF-8 encoding when loading
loader = TextLoader(r"C:\Users\gauri\OneDrive\Desktop\genai\cpptextfile.txt", encoding="utf-8")
finalDocument = loader.load()

# Initialize embeddings
embeddings = HuggingFaceEmbeddings(model_name="sentence-transformers/all-MiniLM-L6-v2")

# Build FAISS index
db = FAISS.from_documents(finalDocument, embeddings)

# Query the FAISS index
query = input("Enter your questions for C++ : ")
docs = db.similarity_search(query, k=3)

# Print results
for i, doc in enumerate(docs, start=1):
    print(f"Result {i}:")
    print(doc.page_content[:300])  # show first 300 characters
    print("---")

```
